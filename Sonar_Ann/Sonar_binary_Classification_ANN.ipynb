```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("sonar.csv",header=None)
```

```python
data=data.sample(frac=1)
data.replace({"M":0,"R":1},inplace=True)
```

```python
data
```

# Data splitting


```python
y_data=data.iloc[:,-1]
y_data
```

```python
x_data=data.iloc[:,:60]
```

```python
x_data
```

```python
x_train=x_data.iloc[:round(len(x_data)*.70),:]
```

```python
y_train=y_data.iloc[:round(len(y_data)*.70)]
```

```python
x_test=x_data.iloc[round(len(x_data)*.30):,:]
```

```python
y_test=y_data.iloc[round(len(x_data)*.30):]
```

```python
from keras import models
from keras import layers
```

```python

network = models.Sequential()
network.add(layers.Dense(100, activation='relu', input_shape=(60,)))
network.add(layers.Dense(150, activation='relu'))
network.add(layers.Dense(250, activation='relu'))
network.add(layers.Dense(150, activation='relu'))
network.add(layers.Dense(100, activation='relu'))
network.add(layers.Dense(1, activation='sigmoid'))
```

```python
network.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
network.fit(x_train,y_train, epochs=16)
```

```python
accuracy = network.evaluate(x_test, y_test)
# print('Accuracy: %.2f' % (accuracy*100))
```
